# <span style="color:#ff69b4; font-weight:bold;">SEE + KMeans Clustering Analysis</span>

---

### <span style="color:#ff69b4;">1. Early Prescriptions Have Greater Variability</span>
- In both medA and medB, early prescriptions (p_number 2, 3, 4) exhibit high variability in event intervals.
- Large spreads in refill durations, with outliers reaching up to 500 days for medA and 300 days for medB.
- This suggests that early refills are irregular, possibly due to patients adjusting to their medication schedules or inconsistencies in initial prescription adherence.

### <span style="color:#ff69b4;">2. Convergence in Later Prescriptions</span>
- As the p_number increases (above 10 for medA, above 15–20 for medB), refill intervals stabilize.
- Boxplots narrow, indicating more predictable refill patterns with fewer extreme outliers.
- Maximum durations decrease, suggesting that patients develop a routine over time.

### <span style="color:#ff69b4;">3.  Distinctions Between medA and medB</span>
- **MedA**
    - Broader refill variation in the early phase.
    - Outliers in refill intervals suggest some patients delay refills significantly.
    - Later refill intervals are shorter, indicating more consistent adherence over time.
- **MedB**
    - Generally shorter refill intervals compared to medA.
    - More frequent prescriptions suggest higher dosage frequency or a stricter regimen.
    - Early refills show more extreme fluctuations, indicating some initial adherence challenges.


# <span style="color:#ff69b4; font-weight:bold;">SEE + DBSCAN Analysis</span>

---

### <span style="color:#ff69b4;">1. Early Prescriptions Show More Outliers and Noise</span>
- DBSCAN identified a significant number of noise points in early refills, particularly in medA.
- Outliers in refill durations exceed 300–500 days, more than K-Means detected.
- The clustering suggests a subset of patients with highly irregular early refill behavior.

### <span style="color:#ff69b4;">2. Later Prescriptions Form Well-Defined Clusters</span>
- DBSCAN was better at grouping consistent refill intervals while ignoring noise.
- For p_numbers above 10–15, clusters became denser and more defined, indicating a transition to routine refill behavior.
- Silhouette scores showed higher clustering quality in later prescriptions than in early ones.

### <span style="color:#ff69b4;">3. Distinctions Between medA and medB</span>
- **Cluster Formation**
    - medA had a higher number of outliers and noise points in the early prescription phase.
    - medB showed more distinct, compact clusters, suggesting better adherence.
- **Refill Interval Trends**
    - medA: DBSCAN found a wider range of refill intervals, supporting the observation that some patients had extreme delays in early refills.
    - medB: DBSCAN grouped refills into shorter, more frequent clusters, reinforcing the idea that medB patients have shorter prescription cycles.
- **Noise Handling**
    - DBSCAN classified a portion of early refills as noise, meaning those prescriptions did not fit into any clear pattern.
    - This suggests a need for further investigation into patient adherence in the early phases.



# <span style="color:#ff69b4; font-weight:bold;">Comparison: SEE + KMeans vs. SEE + DBSCAN</span>

---

### <span style="color:#ff69b4;">1. Early Prescription Variability</span>
- **Variability in early refills**
    - KMeans captures some high-variance prescriptions but forces them into clusters.
    - DBSCAN identifies extreme outliers and classifies them as noise.
- **Outlier handling**
    - Kmeans includes all points in a cluster, even if some are significantly different.
    - DBSCAN filters out noise, making early refill trends more distinct.

### <span style="color:#ff69b4;">2. Stabilization in Later Prescriptions</span>
- **Cluster consistency**
    - KMeans custers become more compact over time, showing stabilized refill patterns.
    - DBSCAN later refills form distinct clusters, showing better-defined refill behavior.
- **Separation of groups**
    - Kmeans may struggle with overlapping refill intervals.
    - DBSCAN forms clearer, density-based clusters, even with overlapping points.

### <span style="color:#ff69b4;">3. Distinctions Between medA and medB</span>
- **medA refill trends**
    - KMeans detects broad clusters but does not separate extreme early refills well.	
    - DBSCAN identifies many early refills as noise, highlighting their irregularity.
- **medB refill trends**
    - Kmeans shows shorter refill intervals but may force some long-interval cases into clusters.
    - DBSCAN clusters are tighter and more distinct, reinforcing frequent refills.
